```python
# import necessary packages
import pandas as pd
import geopandas as gpd
import mapclassify
import folium
import random
import plotly.express as px
from plotly.express import scatter_geo
import warnings
warnings.filterwarnings('ignore')

import matplotlib as mpl
mpl.rcParams['figure.dpi']= 300
```

```python
#Answer 1
#I chose nuclear_power_plants dataset from Kaggle. I chose it because it is a spatial dataset with a manageable number of coulmns. 
#It has missing values, but not as much as it cannot be handled. I did not have a very big dataset, but for this assignment, I did 
#random sampling for a better map. 
#Also, since I am only interested in active nuclear centers, I filtered the dataset so that it only shows the active ones. 
```

```python
#Read my dataset
file_path = 'nuclear_power_plants.csv'
df = gpd.read_file(file_path)
```

```python
df
```

```python
#Filtering closed nuclear centers 
df=df[df["Status"]!="Shutdown"].copy()
df
```

```python
subset_df = df[df['Country'] == 'United States']
subset_df
```

```python
#The columns "Capacity", "Latitude", and "Longitude" were stored in object. for mapping, I needed to coerce them numeric type.
#I also coerced textual columns to string type. 
df['Capacity'] = df['Capacity'].replace('', '0').astype(int)
```

```python
df['Latitude'] = df['Latitude'].replace('', '0.0').astype(float)
```

```python
df['Longitude'] = df['Longitude'].replace('', '0.0').astype(float)
```

```python
df = df.loc[(df['Latitude'] != 0.0) & (df['Longitude'] != 0.0)]
```

```python
df['Status'] = df['Status'].astype(str)
```

```python
df['Country'] = df['Country'].astype(str)
```

```python
df['Name'] = df['Name'].astype(str)
```

```python
#Checking how many unique countries I have
unique_country_count = len(df['Country'].unique())
unique_country_count
```

```python
country_counts = df['Country'].value_counts()

# Print the result
print(country_counts)
```

```python
#Random sampling for a manageable dataset. Made sure that all countries are represented in the sample dataset.

# Replace 'Country' with the actual column name in your dataset
unique_countries = df['Country'].unique()

# Set the number of samples you want to take to 261
num_samples = 261

# Create an empty list to store the random sample indices
sample_indices = []

# Calculate the number of occurrences of each country
country_counts = df['Country'].value_counts().to_dict()

# Randomly sample each country fairly
for country in unique_countries:
    occurrences = country_counts[country]
    sample_size = min(occurrences, num_samples // len(unique_countries))
    sample_indices.extend(random.sample(df[df['Country'] == country].index.tolist(), sample_size))

# Ensure the final sample size matches the desired number
if len(sample_indices) > num_samples:
    sample_indices = sample_indices[:num_samples]

# Select rows from the original DataFrame based on the sampled indices
sample_df = df.loc[sample_indices]

sample_df
```

```python
#Making sure all countries are represented after random sampling 
unique_country_count = len(sample_df['Country'].unique())
unique_country_count
```

```python
#Answer 2
# I am deleting a couple of columns from my dataset. In my analysis, i am interested in seeing the capacity of active nuclear centers 
#across the world. Therefore, "CountryCode" and "ReactorModel" are not as important in the analysis. Plus, in the column status, i 
#only pick active nuclear centers, so I also do not need the columns: "ConstructionStartAt","OperationalFrom","OperationalTo",
#"LastUpdatedAt"and "Source" I only picked active ones and the start and end dates do not matter in my analysis and I am not interested in
#seeing the source either. Additionally, i have Latitude and Longitude in my dataset, so I do not need geometry either. 

columns_to_remove = ["CountryCode", "ReactorModel", "ConstructionStartAt","OperationalFrom","OperationalTo","LastUpdatedAt","Source","geometry"]

#Answer 3
#By removing the above columns, I made a subset of columns I need for aforementioned analysis, to check the capacity of active nuclear centers
#across the world. 
# Remove the specified columns
df = sample_df.drop(columns=columns_to_remove)
df
```

```python
#Answer 4
#From my subset, I chose Capacity, Country, and Name of the nuclear center, and mapped it.

map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
map = folium.Map(location=map_center, zoom_start=5)

for index, row in df.iterrows():
    popup_content = f"""
        <strong>Name:</strong> {row['Name']}<br>
        <strong>Country:</strong> {row['Country']}<br>
        <strong>Capacity:</strong> {row['Capacity']} MW
    """
    folium.Marker(location=[row['Latitude'], row['Longitude'],], 
                  popup=folium.Popup(popup_content, max_width=300)).add_to(map)
map
```
